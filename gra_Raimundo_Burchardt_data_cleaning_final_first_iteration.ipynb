# From here on,  a new dataset will be created with automatically faked generated first and last names and this will be appended to the above cleaned dataset "mentees_mentor_world.csv" to emulate the exact inputs that a user would generate when interacting with the python algorithm and user interface

```python
#creating fake names
```

```python
# through a function and using the faker library we can generate names and first names
```

```python
def random_names(name_type, size):
    """
    Generate n-length ndarray of person names.
    name_type: a string, either first_names or last_names
    """
    names = getattr(Provider, name_type)
    return np.random.choice(names, size=size)
```

```python
size = 109 # size of the dataset
df = pd.DataFrame(columns=['first', 'last']) # creating dataframe with columns  first and last name.
df['first'] = random_names('first_names', size)
df['last'] = random_names('last_names', size)
```

```python
#lowering stirngs of columns to get uniformity across all datasets
df["first"] = df["first"].str.lower()
df["last"] = df["last"].str.lower()
```

#here a new df will be created, where the first 5 letters of the first name [0:4] together with the first 2 of the second name [0:2] will be used to generate the "username", which will be concatenated afterwards.

```python
 faker_df = df.assign(col3=df['first'].str[:4]+df['last'].str[:1])
```

```python
#replace generated col3 with username
faker_df.rename(columns={'col3':'username'}, inplace=True)
#create a new column with password '123' for all users
faker_df['password'] = '123'
```

```python
faker_df = faker_df[['username','password','first','last']]
```
